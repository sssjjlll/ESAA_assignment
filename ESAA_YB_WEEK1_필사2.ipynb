2. DataFrame iloc[] 연산자

--> iloc[]는 위치 기반 인덱싱 방식으로 동작한다. 행과 열 위치를, 0을 출발점으로 하는 세로축, 가로축 좌표 정숫값으로 지정하는 방식

--> iloc[]은 위치 기반 인덱싱만 허용하기 때문에 행과 열의 좌표 위치에 해당하는 값으로 정숫값 또는 정수형의 슬라이싱, 팬시 리스트 값을 입력해줘야 함.

```python
data = {'Name':['Chulmin', 'Eunkyung','Jinwoong', 'Soobeom'], 'Year':[2011,2016,2015,2015], 'Gender': ['Male','Female','Male','Male']}
data_df = pd.DataFrame(data, index=['one','two','three','four'])
data_df
```

```python
data_df.iloc[0,0]
```

--> iloc[]에 DataFrame의 인덱스 값이나 칼러명을 입력하면 오류를 발생시킴

```python
# 아래 코드는 오류를 발생시킵니다.
data_df.iloc[0,'Name']
```

```python
data_df.iloc['one',0]
```

--> iloc[]은 열 위치에 -1을 입력하여 DataFrame의 가장 마지막 열 데이터를 가져오는 데 자주 사용.

* iloc[:,-1]을 하면 맨 마지막 칼럼의 값을 가져오고, iloc[:,:-1]을 하면 처음부터 맨 마지막 칼럼을 제외한 모든 칼럼의 값을 가져오게 됨


```python
print("\n 맨 마지막 칼럼 데이터 [:,-1]\n", data_df.iloc[:,-1])
print("\n 맨 마지막 칼럼을 제외한 모든 데이터 [:,:-1]\n", data_df.iloc[:,:-1])
```

--> iloc[]은 슬라이싱과 팬시 인덱싱은 제공하나 명확한 위치 기반 인덱싱이 사용되어야 하는 제약으로 인해 불린 인덱싱은 제공하지 않음

<br>

3. DataFrame loc[] 연산자

--> loc[]은 명칭 기반으로 데이터를 추출. 행 위치에는 DataFrame 인덱스 값을, 그리고 열 위치에는 칼럼명을 입력해서 loc[인덱스값, 칼럼명]와 같은 형식으로 데이터를 추출.

```python
data_df.loc['one','Name']
```

주의: 명칭 기반이라는 문맥적 의미 때문에 loc[]을 사용 시 행 위치에 DataFrame 인덱스 값이 들어가는 부분을 간과할 수 있음.

```python
# 다음 코드는 오류를 발생시킵니다.
data_df.loc[0,'Name']
```

주의: loc[]에 슬라이싱 기호 ':'을 적용하면 종료 값 -1까지의 범위가 아닌, 종료 값까지 포함하는 것을 의미

--> 이는 명칭 기반 인덱싱의 특성 때문으로, 명칭은 숫자형이 아닐 수 있기 때문에 -1을 할 수 없음

```python
print('위치 기반 iloc slicing\n', data_df.iloc[0:1,0],'\n')
print('명칭 기반 loc slicing\n', data_df.loc['one':'two', 'Name'])
```

**iloc[],loc[]의 특징과 주의할 점 정리**

1. 개별 또는 여러 칼럼 값 전체를 추출하고자 한다면 DataFrame['칼럼명']으로 충분. 하지만 행과 열을 함께 사용하여 데이터를 추출해야 한다면 iloc[]나 loc[] 사용해야 함.

2. DataFrame의 인덱스나 칼럼명으로 데이터에 접근하는 것은 명칭 기반 인덱싱. 0부터 시작하는 행, 열의 위치 좌표에만 의존하는 것이 위치 기반 인덱싱.

3. iloc[]은 위치 기반 인덱싱만 가능. loc[]은 명칭 기반 인덱싱만 가능

4. 명칭 기반 인덱싱에서 슬라이싱을 '시작점:종료점'으로 지정할 때 시작점에서 종료점을 포함한 위치에 있는 데이터를 반환.

###4-10) apply lambda 식으로 데이터 가공

--> 판다스는 apply 함수에 lambda 식을 결합해 DataFrame이나 Series의 레코드별로 데이터를 가공하는 기능을 제공

--> lambda 식은 파이썬에서 함수형 프로그래밍을 지원하기 위해 만들어짐

```python
def get_square(a): return a**2

print('3의 제곱은:', get_square(3))
```

```python
lambda_square = lambda x : x**2
print('3의 제곱은:', lambda_square(3))
```

--> lambda 식을 이용할 때 여러 개의 값을 입력 인자로 사용해야 할 경우에는 보통 map() 함수를 결합해서 사용


```python
a=[1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
```

lambda 식에서 if else 지원할 때 주의:

1) if 절의 경우 if 식보다 반환 값을 먼저 기술해야 함. 이는 lambda 식 ':' 기호의 오른편에 반환 값이 있어야 하기 때문

2) if, else만 지원하고 if, else if, else와 같이 else if는 지원하지 않음

else if를 이용하기 위해서는 else 절을 ()로 내포해 () 내에서 다시 if else 적용해 사용

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성

def get_category(age):
  cat = ''
  if age <= 5: cat = 'Baby'
  elif age <= 12: cat = 'Child'
  elif age <= 18: cat = 'Teenager'
  elif age <= 25: cat = 'Student'
  elif age <= 35: cat = 'Young Adult'
  elif age <= 60: cat = 'Adult'
  else : cat = 'Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
```
